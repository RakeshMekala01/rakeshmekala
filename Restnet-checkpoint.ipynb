```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet152V2

```

```python
# Define paths to your dataset
# train_dir = 'C:/Users/intel2/Desktop/VTPIM01/dataset/train'
# train_dir = 'C:/Users/intel2/Desktop/VTPIM01/dataset/val'
test_dir = 'C:/Users/intel2/Desktop/VTPIM01/dataset/test'
```

```python
# Define image dimensions and batch size
img_height = 224
img_width = 224
batch_size = 32

# Create ImageDataGenerator instances for training, validation, and test sets with data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Flow training, validation, and test images in batches from directories
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'  # Use 'sparse' for sparse categorical labels
)

# val_generator = val_datagen.flow_from_directory(
#     val_dir,
#     target_size=(img_height, img_width),
#     batch_size=batch_size,
#     class_mode='binary'
# )

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.optimizers import Adam

# Define image dimensions and batch size
img_height = 224
img_width = 224
batch_size = 32
num_classes = 2

# Load pre-trained VGG16 model without including the top (fully connected) layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

# Freeze the pre-trained layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom top layers for classification
flatten = Flatten()(base_model.output)
fc1 = Dense(256, activation='relu')(flatten)
output = Dense(num_classes, activation='softmax')(fc1)  # Use 'sigmoid' activation for binary classification


# Combine base model and top layers
model = Model(inputs=base_model.input, outputs=output)

# Compile the model
model.compile(optimizer=Adam(lr=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=2,  # Adjust number of epochs as needed
    # validation_data=val_generator,
    # validation_steps=val_generator.samples // batch_size
)

# Evaluate the model
test_loss, test_acc = model.evaluate(test_generator)
print("Test Accuracy:", test_acc)
```

```python
model.save("vgg16_model.h5")
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, Activation, BatchNormalization, MaxPooling2D, Flatten, Dense, Dropout
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, Activation, BatchNormalization, MaxPooling2D, Flatten, Dense, Dropout

from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Define the CCTAN model
num_classes = 2
def CCTAN(input_shape, num_classes):
    inputs = Input(shape=input_shape)
   

    # Convolutional Layer 1
    conv1 = Conv2D(32, (3, 3), activation='relu')(inputs)
    conv1 = BatchNormalization()(conv1)
    conv1 = Conv2D(32, (3, 3), activation='relu')(conv1)
    conv1 = BatchNormalization()(conv1)
    conv1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    # Convolutional Layer 2
    conv2 = Conv2D(64, (3, 3), activation='relu')(conv1)
    conv2 = BatchNormalization()(conv2)
    conv2 = Conv2D(64, (3, 3), activation='relu')(conv2)
    conv2 = BatchNormalization()(conv2)
    conv2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    # Flatten layer
    flatten = Flatten()(conv2)

    # Fully connected layers
    fc1 = Dense(512, activation='relu')(flatten)
    dropout1 = Dropout(0.5)(fc1)
    fc2 = Dense(256, activation='relu')(dropout1)
    dropout2 = Dropout(0.5)(fc2)

    # Output layer
    outputs = Dense(1, activation='sigmoid')(dropout2)

    # Create model
    model = tf.keras.models.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
from tensorflow.keras.optimizers import Adam


# Build and compile the model
model = CCTAN(input_shape=(img_height, img_width, 3), num_classes=num_classes)
model.compile(optimizer=Adam(lr=0.0001), loss='binary_crossentropy', metrics=['accuracy'])


# Train the model
model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=10,
    validation_data=val_generator,
    validation_steps=val_generator.samples // batch_size
)

# Evaluate the model
model.evaluate(test_generator)
```

```python
# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(test_generator)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
model.save("C2T2Net.h5")
```

```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the saved model
loaded_model = tf.keras.models.load_model('C2T2Net.h5')

# Define a function to preprocess the input image
def preprocess_image(img_path, target_size=(224, 224)):
    img = cv2.imread(img_path)
    img = cv2.resize(img, target_size)
    img = img / 255.0  # Normalize the image
    return img

# Define class labels
class_labels = ['benign', 'malignant']

# Define a function to make predictions using the loaded model
def predict_image(model, img_path):
    img = preprocess_image(img_path)
    img_array = np.expand_dims(img, axis=0)  # Add batch dimension
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction)
    confidence = prediction[0][predicted_class]
    return class_labels[predicted_class], confidence

# Define the path to the input image
input_image_path = 'C:/Users/intel2/Desktop/VTPIM01/dataset/train/malignant/SOB_M_DC-14-2773-400-027.PNG'

# Make predictions using the loaded model
predicted_class, confidence = predict_image(loaded_model, input_image_path)

# Print the predicted class and confidence
print("Predicted Class:", predicted_class)
print("Confidence:", confidence)

# Display the input image
img = cv2.imread(input_image_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
